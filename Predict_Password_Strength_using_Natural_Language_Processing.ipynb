```python
password_tuple=np.array(data)
password_tuple
```

```python
#shuffle data
import random
random.shuffle(password_tuple)
```

```python
# extract data trainig from dataset 
x=[labels[0] for labels in password_tuple]
x
```

```python
#labels y target values
y=[labels[1] for labels in password_tuple]
```

```python
def word_divide_char(inputs):
  character=[]
  for i in inputs:
    character.append(i)
  return character
```

```python
word_divide_char('kyxvufl37')
```

Tfidf explain :

https://medium.com/analytics-vidhya/tf-idf-term-frequency-technique-easiest-explanation-for-text-classification-in-nlp-with-code-8ca3912e58c3

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vectorizer = TfidfVectorizer(tokenizer=word_divide_char)
```

```python
X = vectorizer.fit_transform(x)
```

```python
X.shape
```

```python
vectorizer.get_feature_names()
```

```python
first_document_vector=X[0]
first_document_vector
```

```python
first_document_vector.T.todense()
```

```python
df=pd.DataFrame(first_document_vector.T.todense(),index=vectorizer.get_feature_names(),columns=['TF_IDF'])
df.sort_values(by='TF_IDF',ascending=False)
```

You will use the scikit-learn library to create your models. When coding, this library is written as sklearn, as you will see in the sample code. Scikit-learn is easily the most popular library for modeling the types of data typically stored in DataFrames.

The steps to building and using a model are:

Define: What type of model will it be? A decision tree? Some other type of model? Some other parameters of the model type are specified too.
Fit: Capture patterns from provided data. This is the heart of modeling.
Predict: Just what it sounds like
Evaluate: Determine how accurate the model's predictions are.
Here is an example of defining a decision tree model with scikit-learn and fitting it with the features and target variable.

The scikit-learn library has a function train_test_split to break up the data into two pieces. We'll use some of that data as training data to fit the model, and we'll use the other data as validation data.

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_val,y_train,y_val=train_test_split(X,y,test_size=0.2)
```

```python
X_train.shape
```

```python
from sklearn.linear_model import  LogisticRegression
```

```python
clf=LogisticRegression(random_state=0,multi_class='multinomial')
```

```python
clf.fit(X_train,y_train)
```

```python
dt=np.array(['#543bcd'])
pred=vectorizer.transform(dt)
clf.predict(pred)
```

```python
y_pred =clf.predict(X_val)
y_pred
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
```

```python
cm=confusion_matrix(y_val,y_pred)
print(cm)
print(accuracy_score(y_val,y_pred))
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_val,y_pred))
```
